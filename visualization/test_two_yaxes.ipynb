```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_columns = ['data01', 'data02']
index = ['AAAA', 'BBBB', 'CCCC', 'DDDD', 'EEEE', 'FFFF', 'GGGG']

data_example = pd.DataFrame([[np.sqrt(i+1)*(-1)**(i+1), 1/(i+2)*(-1)**i] for i in range(7)], 
                            index=index, 
                            columns=data_columns)
print(data_example)
```

```python
data_example.reset_index(inplace=True, drop=True)
print(data_example)
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(data_example['data01'], 'r')
ax1.set_xlabel('xlabel')
ax1.set_ylabel('ax1 ylabel', color='r')
ax1.set_ylim()


ax2 = ax1.twinx()
ax2.plot(data_example['data02'], 'b')
ax2.set_ylabel('ax2 ylabel', color='b')
plt.show()
```
