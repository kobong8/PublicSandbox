```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlt
```

```python
plt.rcParams.update({'font.size': 14})
```

```python
np.random.seed(42)
np_data = np.random.rand(100,3)
df = pd.DataFrame(np_data, columns=["col01", "col02", "col03"])

class_num = 3
df["col04"] = np.random.randint(3, size=100)
```

```python
# scatter plot
plt.scatter(df["col01"], df["col02"], c=df["col03"],
            cmap="jet")

plt.xlabel("xlabel")
plt.ylabel("ylabel")

plt.colorbar()
plt.show()
```

```python
# Scatter plot - log scale
plt.scatter(df["col01"], df["col02"], c=df["col03"],
            cmap="jet")
# plt.scatter(df["col01"], df["col02"], c=df["col03"],
#             norm=mlt.colors.LogNorm())

plt.xscale("log")
plt.yscale("log")

plt.xlabel("xlabel")
plt.ylabel("ylabel")

plt.colorbar()
plt.show()
```

```python
# Scatter plot
plt.scatter(df["col01"], df["col02"], c=df["col04"],
            cmap=plt.get_cmap("jet", class_num))

plt.xlabel("xlabel")
plt.ylabel("ylabel")

plt.colorbar(ticks=range(class_num), format="color: %d", label="color")
plt.show()
```
