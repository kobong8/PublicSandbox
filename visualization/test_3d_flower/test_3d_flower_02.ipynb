```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
n = 800
A = 1.995653
B = 1.27689
C = 8
r = np.linspace(start=0, stop=1, num=n)
theta = np.linspace(start=-2, stop=(20 * np.pi), num=n)
[R, THETA] = np.meshgrid(r, theta)
R = R.transpose()
THETA = THETA.transpose()
```

```python
petalNum = 3.6
x = 1 - (1 / 2) * ((5 / 4) * (1 - np.mod(petalNum * THETA, 2 * np.pi)
                              /np.pi) ** 2 - 1/4) ** 2
phi = (np.pi / 2) * np.exp(-THETA / (C * np.pi))
y = A * (R ** 2) * (B * R - 1) ** 2 * np.sin(phi)
R2 = x * (R * np.sin(phi) + y * np.cos(phi))
X = R2 * np.sin(THETA)
Y = R2 * np.cos(THETA)
Z = x * (R * np.cos(phi) - y * np.sin(phi))
```

```python
red_map = np.zeros([10, 3]) 
red_map[:, 0] = np.linspace(1, 0.25, 10)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.YlOrRd)
plt.axis('off')
plt.grid(b=None)
```
