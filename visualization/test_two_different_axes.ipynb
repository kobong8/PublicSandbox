```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_columns = ['data01', 'data02']
index = ['AAAA', 'BBBB', 'CCCC', 'DDDD', 'EEEE', 'FFFF', 'GGGG']

data_example = pd.DataFrame([[np.sqrt(i+1)*(-1)**(i+1), 1/(i+2)*(-1)**i] for i in range(7)], 
                            index=index, 
                            columns=data_columns)
print(data_example)
```

```python
x_axis = np.arange(len(index))

fig, ax1 = plt.subplots()
bar01 = ax1.bar(x_axis - 0.2, data_example['data01'], 0.4, color='red')
ax1.set_ylabel('data01', color='red')
plt.ylim([-3, 3])

ax2 = ax1.twinx()

bar02 = ax2.bar(x_axis + 0.2, data_example['data02'], 0.4, color='blue')
ax2.set_ylabel('data02', color='blue')
plt.ylim([-1, 1])

plt.xticks(x_axis, index)
plt.legend((bar01, bar02), ('data01', 'data02'), loc='upper right')
plt.show()
```

```python
y_axis = np.arange(len(index))

fig, ax1 = plt.subplots()
bar01 = ax1.barh(y_axis - 0.2, data_example['data01'], 0.4, color='red')
plt.xlim([-3, 3])
ax1.set_xlabel('data01', color='r')

ax2 = ax1.twiny()

bar02 = ax2.barh(y_axis + 0.2, data_example['data02'], 0.4, color='blue')
plt.xlim([-1, 1])
ax2.set_xlabel('data02', color='blue')

plt.yticks(y_axis, index)
plt.legend((bar01, bar02), ('data01', 'data02'), loc='upper right')
plt.show()
```
