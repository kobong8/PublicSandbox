```python
import pandas as pd
import random

from typing import List

random.seed(42)
```

- groupby to reorganize dataframe

```python
temp: List = ['A', 'B', 'C']
data: List = [[temp[i % len(temp)], random.random()] for i in range(18)]

df: pd.DataFrame = pd.DataFrame(data, columns=['type', 'value'])
df
```

```python
print(df.columns)
```

```python
type_list = ['A', 'B', 'C']
df_value = pd.DataFrame()
for t in type_list:
    type_value = df[df['type'] == t]
    df_value[t] = type_value.iloc[:, 1]

df_value
```

```python
type_list = ['A', 'B', 'C']
df_value = pd.DataFrame()
for t in type_list:
    type_value = df[df['type'] == t]
    df_value[t] = type_value.iloc[:, 1].values

df_value
```

```python
type_list = ['A', 'B', 'C']
df_value = pd.DataFrame()
for t in type_list:
    type_value = df[df['type'] == t]
    df_value[t] = type_value['value'].values

df_value
```

- groupby for statistical analysis

```python
temp01: List = ["A", "B", "C"]
temp02: List = ["1", "2"]
data: List = [[temp01[i % len(temp01)], temp02[i % len(temp02)], 
               random.random(), random.random()] for i in range(10)]

df: pd.DataFrame = pd.DataFrame(data, columns=["type", "case", "value", "col"])
df
```

```python
df.groupby("type")["value"].groups
```

```python
df.groupby("type")["value"].mean()
```

```python
df.groupby(["type", "case"])["value"].mean()
```

```python
df.groupby("type")[["value", "col"]].mean()
```
