```python
import pandas as pd
import random

from typing import List

random.seed(42)
```

```python
data_merged_cells = pd.read_excel("sample_merged_cells.xlsx")
data_merged_cells
```

```python
data_merged_cells.iloc[0] = data_merged_cells.iloc[0].fillna(method='ffill')
data_merged_cells
```

```python
row_cat: List = list(data_merged_cells.loc[0])
row_var: List = list(data_merged_cells.loc[1])

new_column_name: List = [row_cat[i] + '_' + row_var[i] for i in range(len(row_cat))]
new_column_name
```

```python
data_merged_cells = data_merged_cells.drop([0, 1]).reset_index(drop=True)

data_merged_cells.columns = new_column_name
data_merged_cells
```
