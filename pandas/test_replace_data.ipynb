```python
import pandas as pd
import random

from typing import Dict, List

random.seed(42)
```

```python
temp_num: List[int] = [1, 2]
temp_str: list[str] = ["한국어", "영어", "스페인어"]
data: List = [[random.random(), temp_num[i % len(temp_num)], 
               random.random(), temp_str[i % len(temp_str)],
               random.random()] for i in range(10)]

df: pd.DataFrame = pd.DataFrame(data, columns=['col1', 'col2', 'col3', 'col4', 'col5'])
df.head()
```

```python
df["col6"] = df["col2"]
df.head()
```

```python
# case01 map
df["col2"] = df["col6"]

dict_num: Dict[int, str] = {1: "일", 2: "이"}
df["col2"] = df["col2"].map(dict_num)
df.head()
```

```python
# case02 function
df["col2"] = df["col6"]

def num_to_str(num):
    if num == 1:
        return "일"
    else:
        return "이"

df["col2"] = df["col2"].apply(num_to_str)
df.head()
```

```python
# case03 lambda function
df["col2"] = df["col6"]

df["col2"] = df["col2"].apply(lambda num : "일" if num == 1 else "이")
df.head()
```
