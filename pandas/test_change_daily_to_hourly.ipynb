```python
import pandas as pd
import random

from typing import List

random.seed(42)
```

```python
list_sample: List = [[random.random(), random.random(), random.random(), random.random()] for _ in range(365)]
df_sample: pd.DataFrame = pd.DataFrame(list_sample, columns=['data01', 'data02', 'data03', 'data04'])
df_sample
```

```python
list_new_sample: List = []
for i in range(365):
    for j in range(24):
        list_new_sample.append(df_sample.loc[i].values)

df_new_sample = pd.DataFrame(list_new_sample, columns=df_sample.columns)
df_new_sample
```
