```python
import pandas as pd
from typing import Dict
```

```python
file_path = "sample_replace_data.xlsx"
df = pd.read_excel(file_path)
df["col6"] = df["col2"]
df.head()
```

```python
# case01 map
df["col2"] = df["col6"]

dict_num: Dict[int, str] = {1: "일", 2: "이"}
df["col2"] = df["col2"].map(dict_num)
df.head()
```

```python
# case02 function
df["col2"] = df["col6"]

def num_to_str(num):
    if num == 1:
        return "일"
    else:
        return "이"

df["col2"] = df["col2"].apply(num_to_str)
df.head()
```

```python
# case03 lambda function
df["col2"] = df["col6"]

df["col2"] = df["col2"].apply(lambda num : "일" if num == 1 else "이")
df.head()
```
